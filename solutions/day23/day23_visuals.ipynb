```python
import day23
```

```python
actual_swarm = day23.scan_nano_bots()
```

```python
print("number of bots in range of origin: ", actual_swarm.bots_in_range((0, 0, 0)))
bot_in_range_of_most_others = max(
    actual_swarm.bots,
    key=lambda bot: actual_swarm.bots_in_range(bot.pos)
)
print(
    "max number of bots in range of point (only considering points where a bot is): ",
    actual_swarm.bots_in_range(bot_in_range_of_most_others.pos),  # 815
    "\nThis happens at point: ", bot_in_range_of_most_others.pos  # (60641402, 12088171, 41773347)
)
```
